```python
import numpy as np
import hepunits.units as u
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.colors import Normalize

plt.style.use("xsectron.mplstyle")

from xsectron.literatures import es_cross_section
from xsectron.neutrino import NeutrinoElectron
```

```python
fig = plt.figure(figsize=(8, 7))
# fig.tight_layout()
gs = gridspec.GridSpec(
    1, 1, figure=fig, left=0.15, right=0.95, top=0.9, bottom=0.15, wspace=0.3, hspace=0.3
)
ax = fig.add_subplot(gs[0, 0])

E = np.linspace(1, 40, 31)
ax.plot(E, es_cross_section(E), "-s", lw=2, markersize=7.0, label="COHERENT 2018")
ax.plot(E, NeutrinoElectron.sigma(E * 1e-3), "-s", lw=2, markersize=7.0, label="My calculation")
ax.set_xlim(E[0], E[-1])
ax.set_ylim(1e-45, 1e-38)
ax.set_yscale("log")
ax.grid()
ax.legend()
ax.legend(title=r"$e^-$")
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())
ax2.set_yscale("log")
ax2.set_yticks([])
ax.set_xlabel(r"$\mathrm{Neutrino\ energy}\ E_{\nu}$" + " [MeV]")
ax.set_ylabel(r"$\mathrm{cross\ section}\ \sigma(E_{\nu})$" + " [cm$^2$]")

fig.savefig("es_comparison.svg")

plt.show()
```
